# Her bir nitelik için normalizasyon işlemi


```python
def min_max_normalize(iris_dataset):
  normalized_data= iris_dataset.copy()
  for feature in iris_dataset.columns:
    min_value = iris_dataset[feature].min()
    max_value = iris_dataset[feature].max()
    normalized_data[feature] = (iris_dataset[feature] - min_value) / (max_value - min_value)
  return normalized_data
```

# Entropy ve İnformation gain sınıflandırma yöntemi ile en iyi bölen niteliği bulup ekrana yazma

```python
X = iris_dataset.iloc[:, :-1] # select all rows, all columns except the last one
y = iris_dataset.iloc[:, -1] # select all rows, last column only
import math

def entropy(y):
    counts = y.value_counts()
    entropy = 0
    for count in counts:
        p = count / len(y)
        entropy -= p * math.log(p, 2)
    return entropy

entropy_y = entropy(y)

def information_gain(X, y, feature):
    entropy_y = entropy(y)
    values = X[feature].unique()
    entropy_feature = 0
    for value in values:
        sub_y = y[X[feature] == value]
        p = len(sub_y) / len(y)
        entropy_feature += p * entropy(sub_y)
    return entropy_y - entropy_feature

best_feature = None
best_gain = 0

for feature in X.columns:
    gain = information_gain(X, y, feature)
    if gain > best_gain:
        best_gain = gain
        best_feature = feature

print(f'En iyi bölen nitelik: {best_feature}')
```

#Gini index sınıflandırma yöntemi ile en iyi bölen niteliği bulup ekrana yazma

```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier

# İris veri setini yükleme
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Initialize the Decision Tree classifier
clf = DecisionTreeClassifier()

# Karar Ağacı sınıflandırıcısını başlatma
clf.fit(X, y)

# Özellik önem puanlarını alma
importances = clf.feature_importances_

# Özellik önem puanlarını yazdırma
print(importances)

# En yüksek önem puanına sahip niteliği bulma
best_attribute_index = importances.argmax()

# En iyi niteliği ekrana yazdırma
print(iris.feature_names[best_attribute_index])
```

# Gini index sınıflandırma yöntemi ile en iyi bölen niteliği bulma fakat bu kod da ekrana niteliğin ismi değil, niteliğin index numarası ekrana yazılacak.
#sepal_length=0,  sepal_width=1, petal_length=2,  petal_width=3
#Bu index numaralara göre en iyi nitelik(petal_length) olduğu için ekrana 2 yazacak 

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# iris veri setini yükleme
X, y = load_iris(return_X_y=True)

# Veri setini eğitim ve test setlerine ayırma
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Belirli bir bölünme için gini indeksini hesaplamak için bir fonksiyon tanımlama
def gini_index(groups, classes):
    # ayırma noktasındaki tüm örnekleri sayma
    n_instances = float(sum([len(group) for group in groups]))
  # her grup için toplam ağırlıklı Gini indeksi
    gini = 0.0
    for group in groups:
        size = float(len(group))
        # sıfıra bölmekten kaçınma
        if size == 0:
            continue
        score = 0.0
        # grubu her sınıfın puanına göre puanlama
        for class_val in classes:
            p = [row[-1] for row in group].count(class_val) / size
            score += p * p
        # grup puanını göreli boyutuna göre ağırlıklandırma
        gini += (1.0 - score) * (size / n_instances)
    return gini

# Veri kümesini belirli bir öznitelik ve değere göre bölmek için bir işlev tanımlama
def test_split(index, value, dataset):
    left, right = list(), list()
    for row in dataset:
        if row[index] < value:
            left.append(row)
        else:
            right.append(row)
    return left, right

# Veri kümesi için en iyi bölmeyi bulmak için bir işlev tanımlama
def get_split(dataset):
    class_values = list(set(row[-1] for row in dataset))
    b_index, b_value, b_score, b_groups = 999, 999, 999, None
    for index in range(len(dataset[0])-1):
        for row in dataset:
            groups = test_split(index, row[index], dataset)
            gini = gini_index(groups, class_values)
            if gini < b_score:
                b_index, b_value, b_score, b_groups = index, row[index], gini, groups
    return {'index':b_index, 'value':b_value, 'groups':b_groups}

# iris veri kümesindeki en iyi niteliği bulma
split = get_split(X_train)
print(f'En iyi bölen niteliğin index numarası: {split["index"]}')
```
